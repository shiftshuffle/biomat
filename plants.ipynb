## plant seed population

[pdf](http://sites.ifi.unicamp.br/aguiar/files/2014/10/Leah-Capitulo1.pdf)

\begin{cases}
      p_{n+1}=\alpha \sigma \gamma p_{n}+ + \beta \sigma (1-\alpha) S_{n} , &  \\
      S_{n+1}= \sigma \gamma p_{n}, & 
    \end{cases}

------------
$p_{n+1}=\alpha \sigma \gamma p_{n}+ \beta \sigma^{2} (1-\alpha) \gamma p_{n-1}  $



```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

itr=50






def pops(α,σ,γ,β):
    
    M=np.array([[α*σ*γ, β*σ*(1-α)], [σ*γ, 0]])
        
    pop=[]
    seed=[]
    pop.append(100.0)
    seed.append(0.0)
    current=np.matrix([pop[0],seed[0]])
    
    for i in range(1,itr):
        current=current.dot(M)
        pop.append(current.item(0))
        seed.append( current.item(1))
        
    plt.rcParams['figure.figsize'] = 13, 13
    plt.rcParams['axes.facecolor'] = '111111' 
    plt.grid(color='orange', linestyle='--', linewidth=.2)
    x =np.linspace(0,10,num=itr)
    plt.grid(True)
    plt.plot(x,pop,'r-o',color=(0,.7,.7))
    plt.ylim(0,100)
    plt.xlim(0,10)
    plt.title(u'population des plantes \n  avec des paramètres \n α={0:g} \n σ={1:g} \n γ={2:g} \n β={3:g}'.format(α,σ,γ,β),loc='left')
    plt.title(u'initial seeds=0 \n initial plants=100 \n M=[[α*σ*γ, β*σ*(1-α)], [σ*γ, 0]]',loc='right')
    
    
     
pops(α=.5,σ=.8,γ=2.0,β=.25)
```
